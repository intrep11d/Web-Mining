```python
# Filters a dataframe of accounts according to keywords
tech_keywords = ["tech", "technology", "engineering", "robotics", "stem ", "computer", "coding", "programming", "it,", "informatics", "developer", "software"]

def filter_tech_orgs(df, keywords, column="username"):
    pattern = '|'.join([f"({k.strip().lower()})" for k in keywords])
    mask = df[column].str.lower().str.contains(pattern, na=False)
    return df[mask]

tech_accounts = filter_tech_orgs(usernames_df_copy, tech_keywords)
```

```python
from TikTokApi import TikTokApi
import json

# Initialize the API
api = TikTokApi()

def get_user_details(username):
    try:
        user = api.user(username=username)
        user_info = {
            "username": user.username,
            "bio": user.signature,  # Bio/description
            "followers": user.stats.get("followerCount", 0),
            "following": user.stats.get("followingCount", 0),
            "likes": user.stats.get("heartCount", 0),
            "videos_count": user.stats.get("videoCount", 0),
            "is_verified": user.verified,  # Check for blue checkmark
            "joined": user.create_time  # Approximate join date
        }
        return user_info
    except Exception as e:
        print(f"Error fetching {username}: {e}")
        return None

# Fetch details for each user
user_details = {}
for username in usernames:
    details = get_user_details(username)
    if details:
        user_details[username] = details

# Save to JSON
with open("user_details.json", "w") as f:
    json.dump(user_details, f, indent=4)
```
