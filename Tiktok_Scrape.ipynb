```python
import nest_asyncio
import asyncio
from TikTokApi import TikTokApi
import os
import json

nest_asyncio.apply()

ms_token = os.environ.get("ms_token", None)

async def trending_videos():
    videos_data = []
    async with TikTokApi() as api:
        await api.create_sessions(ms_tokens=[ms_token], num_sessions=1, sleep_after=3, browser=os.getenv("TIKTOK_BROWSER", "chromium"))
        async for video in api.trending.videos(count=5):  # Use 5 for brevity
            # print(json.dumps(video.as_dict, indent=2, ensure_ascii=False))  # Pretty print
            videos_data.append(video.as_dict)
        # with open("trending_videos.json", "w", encoding="utf-8") as f:
        #     json.dump(videos_data, f, indent=2, ensure_ascii=False)
        return videos_data

results = await trending_videos()
```
